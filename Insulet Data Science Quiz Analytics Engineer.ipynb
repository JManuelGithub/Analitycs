```python
import pyodbc
direccion_servidor = 'datasciencechallenge.database.windows.net'
nombre_bd = 'DATA'
nombre_usuario = 'candidate'
password = 'DataForMe123'
try:
    conexion = pyodbc.connect('DRIVER={ODBC Driver 17 for SQL Server};SERVER=' +
                              direccion_servidor+';DATABASE='+nombre_bd+';UID='+nombre_usuario+';PWD=' + password)
    cursor = conexion.cursor()
    # OK! conexión exitosa
except Exception as e:
    # Atrapar error
    print("Ocurrió un error al conectar a SQL Server: ", e)
```

```python
import pandas as pd
sql="Select * FROM training;"
data = pd.read_sql(sql,conexion)
```

```python
data2=pd.concat([pd.get_dummies(data.iloc[:,1]),data.iloc[:,3:]],axis=1)
data2
```

```python
data2=data2.apply(pd.to_numeric)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

matrix = data2.corr(
    method = 'pearson',  
    min_periods = 1      
)
matrix = data2.corr()
sns.heatmap(matrix, annot=False,vmax=1, vmin=-1, center=0, cmap='vlag')
plt.show()
```

```python
matrix = matrix.unstack()
matrix = matrix[(abs(matrix) >= 0.7) & (matrix!=1)]

print(matrix)
```

```python
import numpy as np
data3=data.iloc[:,3:]
data3=data3.apply(pd.to_numeric)
matrix = data3.corr(
    method = 'pearson', 
    min_periods = 1      
)
mask = np.triu(np.ones_like(matrix, dtype=bool))
sns.heatmap(matrix, annot=False,vmax=1, vmin=-1, center=0, cmap='vlag', mask=mask)
plt.show()
```

```python
matrix = matrix.unstack()
matrix = matrix[(abs(matrix) >= 0.7) & (matrix!=1)]
print(matrix)
```
